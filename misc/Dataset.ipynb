```python
dst = ProcessedCsvDataset(root_dir='../data', split='many')
print(dst.train_feature.shape, dst.train_label.shape)
print(dst.test_feature.shape, dst.test_label.shape)
```

```python
hist_bin = np.linspace(-25, 25, 50)
bin_center = (hist_bin[1:] + hist_bin[:-1]) / 2
train_hist, _ = np.histogram(dst.train_label, bins=hist_bin)
test_hist, _ = np.histogram(dst.test_label, bins=hist_bin)
train_hist = train_hist / train_hist.sum()
test_hist = test_hist / test_hist.sum()
plt.plot(bin_center, train_hist)
plt.plot(bin_center, test_hist)
```
